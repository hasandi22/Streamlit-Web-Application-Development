Handle missing values appropriately

```python
# Checking for missing values
print("Missing values before handling:")
print(df.isnull().sum())
```

```python
# Create a copy of the original DataFrame so we don't overwrite it
df_filled = df.copy()
```

```python
# Replace '?' placeholders with NaN (some datasets use '?' for missing values)
import numpy as np
df_filled.replace('?', np.nan, inplace=True)
```

```python
# Convert columns to numeric where possible (non-numeric values become NaN)
for col in df_filled.columns:
    try:
        df_filled[col] = pd.to_numeric(df_filled[col], errors='coerce')
    except:
        pass  # If conversion fails entirely, skip this column
```

```python
# Median for numeric columns
df_filled[col] = df_filled[col].fillna(df_filled[col].median())

# Mode for categorical columns
df_filled[col] = df_filled[col].fillna(df_filled[col].mode()[0])
```

```python
# Print the count of missing values after handling to verify
print("Missing values after handling:")
print(df_filled.isnull().sum())
```

```python
# Loop through each column
for col in df_filled.columns:
    if df_filled[col].isnull().sum() > 0:  # Only if missing values exist
        if pd.api.types.is_numeric_dtype(df_filled[col]):
            # Fill numeric with median
            df_filled[col] = df_filled[col].fillna(df_filled[col].median())
        else:
            # Fill categorical with mode (most frequent value)
            mode_value = df_filled[col].mode()[0] if not df_filled[col].mode().empty else "Unknown"
            df_filled[col] = df_filled[col].fillna(mode_value)
```

```python
# Verify again
print(df_filled.isnull().sum())
```

```python
# Loop through all columns in the dataframe df_filled
for col in df_filled.columns:
    # Check if the current column has any missing values
    if df_filled[col].isnull().sum() > 0:

        # If the column is numeric type
        if pd.api.types.is_numeric_dtype(df_filled[col]):

            # Check if there is at least one non-missing value
            if df_filled[col].notnull().any():
                # Fill missing values with the median of the column
                df_filled[col] = df_filled[col].fillna(df_filled[col].median())
            else:
                # If the entire column is missing, fill all with 0 (default numeric value)
                df_filled[col] = df_filled[col].fillna(0)

        # If the column is non-numeric (categorical)
        else:
            # Check if there is at least one non-missing value
            if df_filled[col].notnull().any():
                # Fill missing values with the most frequent value (mode) of the column
                df_filled[col] = df_filled[col].fillna(df_filled[col].mode()[0])
            else:
                # If the entire column is missing, fill all with a placeholder string "Unknown"
                df_filled[col] = df_filled[col].fillna("Unknown")
```

```python
# Print the count of missing values in each column after filling
print(df_filled.isnull().sum())
```

Feature Enginnering

```python
#to normalize/standardize numerical features
from sklearn.preprocessing import StandardScaler
```

```python
#Copy of the original filled dataset to avoid modifying it directly
df_fe = df_filled.copy()
```

```python
#Remove ID column, as it doesn't carry predictive information
if 'id' in df_fe.columns:
    df_fe.drop('id', axis=1, inplace=True)
```

```python
#Encode categorical variables into dummy/indicator variables
#These are columns with discrete categories represented as numbers
categorical_cols = ['sex', 'cp', 'restecg', 'slope', 'thal', 'dataset', 'exang', 'fbs', 'ca']
df_fe = pd.get_dummies(df_fe, columns=categorical_cols, drop_first=True)
# drop_first=True avoids the dummy variable trap (multicollinearity)
```

```python
# Creating interaction and ratio features on un-scaled data

# Cholesterol per age
df_fe['chol_per_age'] = df_filled['chol'] / (df_filled['age'] + 1e-6)

# Heart rate reserve
df_fe['hr_reserve'] = df_filled['thalch'] / (220 - df_filled['age'] + 1e-6)

# Count of major vessels colored (sum of ca one-hot columns)
df_fe['ca_count'] = df_fe[['ca_1.0', 'ca_2.0', 'ca_3.0']].sum(axis=1)

# Heart rate per age ratio
df_fe['hr_per_age'] = df_filled['thalch'] / (df_filled['age'] + 1e-6)
```

```python
#Scale all continuous/numeric columns including the new ones
continuous_cols = ['age', 'trestbps', 'chol', 'thalch', 'oldpeak', 'chol_per_age', 'hr_reserve', 'hr_per_age']
scaler = StandardScaler()
df_fe[continuous_cols] = scaler.fit_transform(df_fe[continuous_cols])
```

```python
# Convert any boolean columns to integers
bool_cols = df_fe.select_dtypes(include='bool').columns
df_fe[bool_cols] = df_fe[bool_cols].astype(int)
```

```python
#Scale continuous variables for better model performance
#StandardScaler transforms them to have mean 0 and standard deviation 1
continuous_cols = ['age', 'trestbps', 'chol', 'thalch', 'oldpeak']
scaler = StandardScaler()
df_fe[continuous_cols] = scaler.fit_transform(df_fe[continuous_cols])
```

```python
#Shape of the dataset after transformations
print(f"Shape after feature engineering: {df_fe.shape}")
```

Random Forest hyperparameter tuning

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
#Defining parameter grid for RandomizedSearch
param_dist = {
    'n_estimators': [50, 100, 200, 300],
    'max_depth': [None, 5, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],
    'class_weight': ['balanced'],
}
```

```python
#Initializing Random Forest
rf = RandomForestClassifier(random_state=42)
```

```python
#Setting up RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=20,  # Number of random parameter sets to try
    scoring='accuracy',
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
```

```python
# Fit to training data
random_search.fit(X_train, y_train)

print("Best parameters found:", random_search.best_params_)
```

```python
# Get the best Random Forest model found by hyperparameter tuning
best_rf = random_search.best_estimator_
# Use the tuned model to predict on the test set
y_pred = best_rf.predict(X_test)
```

```python
# Printing the accuracy score of the tuned model
print("Tuned Random Forest Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#Printing detailed classification metrics (precision, recall, f1-score)
print("Classification Report:\n", classification_report(y_test, y_pred))
```

cross-validation for model evaluation

```python
#Initializing Random Forest with tuned hyperparameters
best_rf = RandomForestClassifier(
    n_estimators=300,           # Number of trees in the forest
    min_samples_split=10,       # Min samples required to split a node
    min_samples_leaf=2,         # Min samples required at a leaf node
    max_features='log2',        # Number of features to consider at each split
    max_depth=20,               # Max depth of each tree
    random_state=42             # Seed for reproducibility
)
```

```python
#Evaluating the model using cross-validation and accuracy metric
#n_jobs=-1 means use all CPU cores for faster computation
cv_scores = cross_val_score(best_rf, X, y, cv=skf, scoring='accuracy', n_jobs=-1)
```

```python
#Printing the individual fold accuracies, mean accuracy, and standard deviation
print(f"Cross-validation accuracies: {cv_scores}")
print(f"Mean CV accuracy: {np.mean(cv_scores):.4f}")
print(f"Standard deviation: {np.std(cv_scores):.4f}")
```

SMOTE Usage

```python
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
#Apply SMOTE on training data
#data balancing method used when the dataset is imbalanced
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
```

```python
#Set up XGBoost classifier
xgb_clf = XGBClassifier(random_state=42, eval_metric='mlogloss')
```

```python
#Define hyperparameter grid for tuning
param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0],
    'min_child_weight': [1, 3, 5]
}
```

```python
#Randomized search with 5-fold CV
random_search = RandomizedSearchCV(
    xgb_clf,
    param_distributions=param_dist,
    n_iter=20,
    scoring='accuracy',
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
```

```python
#Fit random search on resampled training data
random_search.fit(X_train_res, y_train_res)

# Save X_test
X_test.to_csv("X_train_res.csv", index=False)

# Save y_test
y_test.to_csv("y_train_res.csv", index=False)

print("Test data saved as X_train_res.csv and y_train_res.csv in the project folder.")
```
